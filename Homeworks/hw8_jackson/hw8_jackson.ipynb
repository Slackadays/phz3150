```python
print( 'Problem 2:' )

class graduation_presents:
    a = 0
    b = 0
    c = 0
    number_of_gifts = 0
    wrapping_paper_available = 0

    def __init__(self, a, b, c, number_of_gifts, wrapping_paper_available):
        self.a = a
        self.b = b
        self.c = c
        self.number_of_gifts = number_of_gifts
        self.wrapping_paper_available = wrapping_paper_available

    def surface_area(self) -> float:
        return 2 * (self.a * self.b + self.a * self.c + self.b * self.c)

    def volume(self) -> float:
        return self.a * self.b * self.c

    def enough_paper(self) -> bool:
        return self.surface_area() <= self.wrapping_paper_available

    def fits(self) -> bool:
        average_gift_volume = 25
        return self.volume() / average_gift_volume >= self.number_of_gifts
        
    def print_gift(self):
        if self.enough_paper():
            print( 'You have enough wrapping paper for these gifts!' )
        else:
            print( 'You don\'t have enough wrapping paper for these gifts.' )
        if self.fits():
            print( 'All these gifts will fit!' )
        else:
            print( 'Not all these gifts will fit.' )
        if not self.enough_paper() and not self.fits():
            print( 'Sorry!' )

try_one = graduation_presents(10, 10, 10, 15, 800)
try_one.print_gift()    
try_two = graduation_presents(10, 10, 10, 55, 1000)
try_two.print_gift()
try_three = graduation_presents(100, 20, 10, 55, 300)
try_three.print_gift()
```
