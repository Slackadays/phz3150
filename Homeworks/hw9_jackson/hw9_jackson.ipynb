```python
print( 'Problem 3:' )

from hw9_support_funct import piston

import matplotlib.pyplot as plt

V = np.linspace(1, 0.1, 40)

PVT = piston(V, 1, 1, 300, 1.4)

P = [PVT[i][0] for i in range(len(PVT))] # extract P, V, and T from PVT
V = [PVT[i][1] for i in range(len(PVT))]
T = [PVT[i][2] for i in range(len(PVT))]

plt.rcParams['figure.dpi'] = 250 # increase matplotlib resolution

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

plt.plot(P, V, T)
plt.title('Temperature (K) vs. Pressure (atm) and Volume (liters)')
ax.set_xlabel('Pressure (atm)')
ax.set_ylabel('Volume (liters)')
ax.set_zlabel('Temperature (K)')
plt.title('Temperature (K) vs. Pressure (atm) and Volume (liters)')
```
